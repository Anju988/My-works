```python
%%writefile iris_net.cpp

#include <iostream>
#include <torch/torch.h>
#include <fstream>
#include <sstream>
#include <vector>
#include <string>
#include <exception>
#include <stdexcept>
#include <unordered_map>
#include <algorithm> // For std::remove_if

struct IrisDataset : torch::data::Dataset<IrisDataset> {
    torch::Tensor features, labels;
    std::unordered_map<std::string, int64_t> label_map = {
        {"Setosa", 0},
        {"Versicolor", 1},
        {"Virginica", 2}
    };

    IrisDataset(const std::string& file_path) {
        std::ifstream file(file_path);
        if (!file.is_open()) {
            throw std::runtime_error("Could not open file: " + file_path);
        }

        std::string line, cell;
        std::vector<std::vector<float>> data;
        std::vector<int64_t> labels_list;

        std::getline(file, line); // skip header

        while (std::getline(file, line)) {
            std::stringstream line_stream(line);
            std::vector<float> row;

            try {
                for (int i = 0; i < 4; ++i) { // Read first 4 columns as floats
                    std::getline(line_stream, cell, ',');
                    row.push_back(std::stof(cell));
                }

                // Read the label (string), trim it and convert it to an integer
                std::getline(line_stream, cell, ',');
                cell.erase(std::remove_if(cell.begin(), cell.end(), [](unsigned char c) {
                    return isspace(c) || c == '\"';
                }), cell.end());
                labels_list.push_back(label_map.at(cell));
                data.push_back(row);
            } catch (const std::invalid_argument& e) {
                std::cerr << "Invalid argument: " << cell << " in line: " << line << std::endl;
                throw;
            } catch (const std::out_of_range& e) {
                std::cerr << "Out of range: " << cell << " in line: " << line << std::endl;
                throw;
            } catch (const std::exception& e) {
                std::cerr << "Exception: " << e.what() << " in line: " << line << std::endl;
                throw;
            }
        }

        std::vector<float> flat_data;
        for (auto& row : data) {
            flat_data.insert(flat_data.end(), row.begin(), row.end());
        }

        features = torch::tensor(flat_data).reshape({static_cast<long>(data.size()), 4});
        labels = torch::tensor(labels_list);
    }

    torch::data::Example<> get(size_t index) override {
        return {features[index], labels[index]};
    }

    torch::optional<size_t> size() const override {
        return features.size(0);
    }
};

struct Net : torch::nn::Module {
    torch::nn::Linear input{nullptr}, hidden_1{nullptr}, output{nullptr};

    Net() {
        input = register_module("input", torch::nn::Linear(4, 16));
        hidden_1 = register_module("hidden_1", torch::nn::Linear(16, 16));
        output = register_module("output", torch::nn::Linear(16, 3));
    }

    torch::Tensor forward(torch::Tensor x) {
        x = torch::relu(input->forward(x));
        x = torch::relu(hidden_1->forward(x));
        return output->forward(x);
    }
};
```

```python
%%writefile main.cpp
#include <iostream>
#include <fstream>
#include <vector>
#include <exception>
#include <stdexcept>
#include "iris_net.cpp" // Ensure this file is correctly implemented and in the same directory
#include <torch/torch.h> // Include PyTorch C++ API

int main() {
    try {
        // Load dataset
        auto dataset = IrisDataset("/content/iris.csv").map(torch::data::transforms::Stack<>());

        // Verify dataset size
        auto dataset_size = dataset.size().value();
        std::cout << "Dataset size: " << dataset_size << std::endl;

        // If the dataset is empty, exit
        if (dataset_size == 0) {
            std::cerr << "Dataset is empty. Please check the dataset file." << std::endl;
            return 1;
        }

        // Define DataLoaderOptions with the desired parameters
        torch::data::DataLoaderOptions options;
        options.batch_size(12);
        options.workers(2);  // Adjust according to your environment

        // Create data_loader using make_data_loader
        auto data_loader = torch::data::make_data_loader(
            std::move(dataset),
            options
        );

        // Calculate number of batches
        int num_batches = std::distance(data_loader->begin(), data_loader->end());
        std::cout << "Number of batches: " << num_batches << std::endl;

        // If there are no batches, exit
        if (num_batches == 0) {
            std::cerr << "No batches created. Please check the DataLoader and dataset." << std::endl;
            return 1;
        }

        // Create test dataset and test_loader (assuming the same dataset)
        auto test_dataset = IrisDataset("/content/iris.csv").map(torch::data::transforms::Stack<>());
        auto test_loader = torch::data::make_data_loader(
            std::move(test_dataset),
            torch::data::DataLoaderOptions().batch_size(30)
        );

        // Initialize model and optimizer
        Net model;
        torch::optim::Adam optimizer(model.parameters(), torch::optim::AdamOptions(0.01));

        int num_epochs = 200;
        std::vector<float> train_accuracies, test_accuracies;

        for (int epoch = 0; epoch < num_epochs; ++epoch) {
            model.train();
            float train_correct = 0.0;
            int train_total = 0;

            for (auto& batch : *data_loader) {
                auto data = batch.data, target = batch.target;
                optimizer.zero_grad();
                auto output = model.forward(data);
                auto loss = torch::nn::functional::cross_entropy(output, target);
                loss.backward();
                optimizer.step();

                auto pred = output.argmax(1);
                train_correct += pred.eq(target).sum().template item<float>();
                train_total += target.size(0);
            }

            // Ensure train_total is greater than 0 before calculating accuracy
            if (train_total > 0) {
                train_accuracies.push_back(100.0 * train_correct / train_total);
            } else {
                std::cerr << "Warning: train_total is zero for epoch " << epoch << std::endl;
                continue;  // Skip calculating accuracy for this epoch
            }

            model.eval();
            torch::NoGradGuard no_grad;
            float test_correct = 0.0;
            int test_total = 0;

            for (auto& batch : *test_loader) {
                auto data = batch.data, target = batch.target;
                auto output = model.forward(data);
                auto pred = output.argmax(1);
                test_correct += pred.eq(target).sum().template item<float>();
                test_total += target.size(0);
            }

            // Ensure test_total is greater than 0 before calculating accuracy
            if (test_total > 0) {
                test_accuracies.push_back(100.0 * test_correct / test_total);
            } else {
                std::cerr << "Warning: test_total is zero for epoch " << epoch << std::endl;
                continue;  // Skip calculating accuracy for this epoch
            }

            std::cout << "Epoch " << epoch << " - Train Accuracy: " << train_accuracies.back() << ", Test Accuracy: " << test_accuracies.back() << std::endl;
        }

        // Save accuracies to a file
        std::ofstream outfile("/content/accuracies.txt"); // Adjust path as needed
        if (!outfile.is_open()) {
            std::cerr << "Error opening file accuracies.txt" << std::endl;
            return 1;
        }

        outfile << "Train Accuracies\n";
        for (const auto& acc : train_accuracies) {
            outfile << acc << "\n";
        }
        outfile << "Test Accuracies\n";
        for (const auto& acc : test_accuracies) {
            outfile << acc << "\n";
        }
        outfile.close();

        std::cout << "Accuracies saved to accuracies.txt" << std::endl;
    } catch (const std::exception& e) {
        std::cerr << "Exception: " << e.what() << std::endl;
        return 1;
    }

    return 0;
}
```

```python
import matplotlib.pyplot as plt

train_accuracies = []
test_accuracies = []
reading_train = False
reading_test = False

with open("accuracies.txt", "r") as f:
    for line in f:
        if line.strip() == "Train Accuracies":
            reading_train = True
            reading_test = False
            continue
        elif line.strip() == "Test Accuracies":
            reading_train = False
            reading_test = True
            continue

        if reading_train:
            train_accuracies.append(float(line.strip()))
        elif reading_test:
            test_accuracies.append(float(line.strip()))

plt.figure()
plt.subplot(2, 1, 1)
plt.plot(train_accuracies)
plt.xlabel("Epoch")
plt.ylabel("Training Accuracy")

plt.subplot(2, 1, 2)
plt.plot(test_accuracies)
plt.xlabel("Epoch")
plt.ylabel("Test Accuracy")


plt.show()
```

```python
%%writefile plot_accuracies.cpp
#include <iostream>
#include <vector>
#include "matplotlibcpp.h"

namespace plt = matplotlibcpp;

int main() {
    // Sample data
    std::vector<double> train_accuracies = {0.1, 0.4, 0.5, 0.7, 0.8, 0.9};
    std::vector<double> test_accuracies = {0.2, 0.3, 0.45, 0.6, 0.7, 0.75};

    // Create a figure with tight layout
    plt::figure_size(800, 600);

    // Create the first subplot
    plt::subplot2grid(2, 1, 0, 0); // 2 rows, 1 column, start at (0,0)
    plt::plot(train_accuracies);
    plt::xlabel("Epoch");
    plt::ylabel("Training accuracy");

    // Create the second subplot
    plt::subplot2grid(2, 1, 1, 0); // 2 rows, 1 column, start at (1,0)
    plt::plot(test_accuracies);
    plt::xlabel("Epoch");
    plt::ylabel("Test accuracy");

    // Save plot to a file
    plt::save("plot.png");

    // Show plot
    plt::show();

    return 0;
}
```
